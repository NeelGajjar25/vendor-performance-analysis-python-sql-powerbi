```python
def format_dollars(value):
    if value >= 1_000_000:
        return f"{value / 1_000_000:.2f}M"
    elif value >= 1_000:
        return f"{value / 1_000:.2f}K"
    else:
        return str(value)
```

```python
def confidence_interval(data, confidence=0.95):
    mean_val = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(len(data)) # Standard error
    t_critical = stats.t.ppf((1 + confidence) / 2, df=len(data) - 1)
    margin_of_error = t_critical * std_err
    return mean_val, mean_val - margin_of_error, mean_val + margin_of_error
```

```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns

# =================================================================
# 1. Define the Confidence Interval Function
# =================================================================
def confidence_interval(data, confidence=0.95):
    """Calculates the mean and confidence interval for a 1D dataset."""
    # Ensure data is numpy array for robust calculations
    data_array = np.asarray(data) 
    
    # Handle the case where the data is empty or has only one element (prevents errors)
    if len(data_array) <= 1:
        return np.nan, np.nan, np.nan

    mean_val = np.mean(data_array)
    # ddof=1 for sample standard deviation
    std_err = np.std(data_array, ddof=1) / np.sqrt(len(data_array)) 
    
    # Calculate t-critical value
    t_critical = stats.t.ppf((1 + confidence) / 2, df=len(data_array) - 1)
    
    margin_of_error = t_critical * std_err
    
    # Returns: Mean, Lower Bound, Upper Bound
    return mean_val, mean_val - margin_of_error, mean_val + margin_of_error

# =================================================================
# 2. Data Preparation & Cleaning (Crucial Step)
#    Assumes 'top_vendors' and 'low_vendors' DataFrames already exist.
# =================================================================

# Select the 'ProfitMargin' column (creating a Series) and drop NaNs.
# NOTE: If 'ProfitMargin' still contains non-numeric strings, you must
# perform pd.to_numeric() conversion before this step.
top_vendors_margin = top_vendors['ProfitMargin'].dropna()
low_vendors_margin = low_vendors['ProfitMargin'].dropna()

# =================================================================
# 3. Calculate Confidence Intervals
# =================================================================
top_mean, top_lower, top_upper = confidence_interval(top_vendors_margin)
low_mean, low_lower, low_upper = confidence_interval(low_vendors_margin)

print(f"Top Vendors 95% CI: ({top_lower:.2f}, {top_upper:.2f}), Mean: {top_mean:.2f}")
print(f"Low Vendors 95% CI: ({low_lower:.2f}, {low_upper:.2f}), Mean: {low_mean:.2f}")

# =================================================================
# 4. Plotting the Comparison Histogram
# =================================================================
plt.figure(figsize=(12, 6))

# Top Vendors Plot (uses the clean 'top_vendors_margin' Series)
sns.histplot(top_vendors_margin, kde=True, color="blue", bins=30, alpha=0.5, label="Top Vendors")
plt.axvline(top_lower, color="blue", linestyle="--", label=f"Top Lower: {top_lower:.2f}")
plt.axvline(top_upper, color="blue", linestyle="--", label=f"Top Upper: {top_upper:.2f}")
plt.axvline(top_mean, color="blue", linestyle="-", label=f"Top Mean: {top_mean:.2f}")

# Low Vendors Plot (uses the clean 'low_vendors_margin' Series)
sns.histplot(low_vendors_margin, kde=True, color="red", bins=30, alpha=0.5, label="Low Vendors")
plt.axvline(low_lower, color="red", linestyle="--", label=f"Low Lower: {low_lower:.2f}")
plt.axvline(low_upper, color="red", linestyle="--", label=f"Low Upper: {low_upper:.2f}")
plt.axvline(low_mean, color="red", linestyle="-", label=f"Low Mean: {low_mean:.2f}")

# Finalize Plot
plt.title("Confidence Interval Comparison: Top vs. Low Vendors (Profit Margin)")
plt.xlabel("Profit Margin (%)")
plt.ylabel("Frequency")

# Set X-axis limits to exclude extreme outliers and zoom in on the main distribution
# Adjust these values if your central data range is different!
plt.xlim(-50, 100) 

plt.legend()
plt.grid(True)
plt.show()
```

```python
from scipy.stats import ttest_ind
top_threshold = df["TotalSalesDollars"].quantile(0.75)
low_threshold = df["TotalSalesDollars"].quantile(0.25)

# Filter the ProfitMargin Series for top and low vendors
top_vendors = df[df["TotalSalesDollars"] >= top_threshold]["ProfitMargin"].dropna()
low_vendors = df[df["TotalSalesDollars"] <= low_threshold]["ProfitMargin"].dropna()

# Perform Two-Sample T-Test
t_stat, p_value = ttest_ind(top_vendors, low_vendors, equal_var=False)

# Print results
print(f"t-Statistic: {t_stat:.4f}, P-Value: {p_value:.4f}")

if p_value < 0.05:
    print("Reject H: There is a significant difference in profit margins between top and low-performing vendors.")
else:
    print("Fail to Reject H: No significant difference in profit margins.")
```
